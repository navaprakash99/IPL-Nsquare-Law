```python
import numpy as np
import pandas as pd
```

```python
ipl_data=pd.read_csv('ipl_f.csv')
```

```python
ipl_data
```

```python
ipl_data=ipl_data[['batsman','bowler','batsman_runs','is_wicket']]
```

```python
ipl_batsman=ipl_data[['batsman','batsman_runs']].groupby(['batsman'])['batsman_runs'].count()
```

```python
ipl_batsman=ipl_batsman.to_dict()
```

```python
ipl_bats=ipl_data[['batsman','batsman_runs']].groupby(['batsman','batsman_runs'])['batsman_runs'].count()
```

```python
ipl_bat_ratio=dict()
```

```python
for batsman in list(ipl_batsman.keys()):
  ipl_bat_ratio[batsman]=dict()
  for i in np.arange(7):
    try:
      ipl_bat_ratio[batsman][i]=np.square(ipl_bats[batsman][i]/ipl_batsman[batsman])
    except:
      ipl_bat_ratio[batsman][i]=0.0
```
